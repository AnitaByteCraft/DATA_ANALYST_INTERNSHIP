# Identified missing values using .isnull().sum().**bold text**

```python
df.shape
```

```python
df.isnull().sum()
```

Dropped Cabin column due to excessive missing data.**bold text**

```python
df.drop(columns=['Cabin'], inplace=True, errors='ignore')


```

Filled numerical missing values (Age) using median.**bold text**

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

Filled categorical missing values (Embarked) using mode.**bold text**

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

Removed duplicate records to avoid biased analysis.**bold text**

```python
before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]

print("Rows before:", before)
print("Rows after:", after)
```

Converted Survived column to categorical datatype.**bold text**

```python
df['Survived'] = df['Survived'].astype('category')
```

Created new column Age_Group to demonstrate data transformation.**bold text**

```python
def age_group(age):
    if age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df['Age_Group'] = df['Age'].apply(age_group)
```
